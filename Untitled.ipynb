```python
from sklearn import linear_model
REG_MODEL = linear_model.LinearRegression()
def linear_regression_value(data_list):
    print(data_list)
    data_size = data_list.size
    x_axis = np.array([range(data_list.size)]).T + 1
    REG_MODEL.fit(x_axis, np.array(data_list))
    return REG_MODEL.predict(np.array(data_size))[0]
```
